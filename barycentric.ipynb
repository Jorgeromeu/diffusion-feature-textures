```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.tri import Triangulation
from pytorch3d.ops import interpolate_face_attributes

triangle = np.array([[0.15, 0.2],
                     [0.1, 0.8],
                     [0.9, 0.7]])

triangulation = Triangulation(x=triangle[:, 0], y=triangle[:, 1], triangles=[[0,1,2]])

res = 100

raster = np.zeros(shape=(res, res))

x, y = np.meshgrid(np.linspace(0, 1, res), np.linspace(0, 1, res))


plt.imshow(raster, extent=[0, 1, 1, 0])
plt.triplot(triangulation, marker='.')
plt.axis('equal')
```
