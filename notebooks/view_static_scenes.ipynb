```python
from text3d2video.artifacts.static_scene import StaticSceneArtifact
from text3d2video.ipython_utils import display_ims
from text3d2video.rendering import render_depth_map

scenes = [
    'booty_static',
    'rumba_static',
    'handstand_back_static',
    'flair_open_static',
    'flair_closed_static'
]

depths = []
for scene in scenes:
    scene = StaticSceneArtifact.from_wandb_artifact_tag(f'{scene}:latest')
    cam, mesh = scene.load_scene()
    depth = render_depth_map(mesh, cam)
    depths.append(depth[0])

display_ims(depths)
```
